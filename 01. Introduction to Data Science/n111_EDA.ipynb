# EDA (Exploratory Data Analysis)


## EDA란 데이터 분석에 있어서 매우 중요한, 초기 분석의 단계를 의미하며
* 시각화 같은 도구를 통해서 패턴을 발견하거나
* 데이터의 특이성을 확인하거나
* 통계와 그래픽 (시각적 표현)을 통해서 가설을 검정하는 과정 등을 포함

<br>

## EDA의 방법은 크게 2가지 (Graphic, Non_Graphic) 으로 나눠질 수 있으며
* Graphic: 차트 혹은 그림 등을 이용하여 데이터를 확인하는 방법
* Non-Graphic: 그래픽적인 요소를 사용하지 않는 방법으로, 주로 Summary Statistics를 통해 데이터를 확인하는 방법


## 동시에 EDA의 타겟 데이터 또한 2가지 (Univariate, Multi-variate)로 나눠진다.
* Multi-variate: 여러 변수들간의 관계를 보는 것이 주요 목적

<br>

# Uni - Non Graphic
Sample Data의 Distribution을 확인하는 것이 주목적
Numeric data의 경우, summary statistics를 제일 많이 사용하는데 이에는
* Center (Mean, Median, Mod)
* Spread (Variance, SD, IQR, Range)
* Modality (Peak)
* Shape (Tail, Skewnewss, Kurtosis)
* Outliers 등을 확인

Categorical data의 경우 occurence, frequency, tabulation등을 할 수 있다.  
<br>

# Uni - Graphic  
Histogram 혹은 Pie chart, Stem-leaf plot, Boxplot, QQplot 등을 사용한다.
그러나 만약 값들이 너무 다양하다면, Binning, Tabultation 등을 활용할 수도 있음

# QQPlot

데이터의 분포와 이론상 분포가 잘 일치하는가를 확인할 수 있는 방법

<br>

# Multi - Non Graphic
* Relationship을 보는 것이 주된 목표
* Cross-Tabulation
* Cross-Statistics (Correlation, Covariance)
등을 사용

Categorical 데이터는 Cross-Tabulation을 적용할 수 있으며, Numerical Features들의 경우 Cross-Statistics를 통해 EDA가 가능 (heat map 처럼)

<br>

# Multi - Graphic
* Category & Numeric: Boxplots, Stacked bar, Parallel Coordinate, Heatmap
* Numeric & Numeric: Scatter Plot
