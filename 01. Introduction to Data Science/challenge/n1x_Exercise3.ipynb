## 1. $\vec{c} \cdot \vec{d}$ 를 구해보세요. 이 두 개의 vector는 orthogonal 관계 (직교)에 있나요? 왜 그런가요? 왜 아닌가요?

\begin{align}
\vec{c} = \begin{bmatrix}3 & 7 & -2 & 12\end{bmatrix}
\qquad
\vec{d} = \begin{bmatrix}9 & -7 & 4 & 6\end{bmatrix}
\end{align}

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
c = np.array([3,7,-2,12])
d = np.array([9,-7,4,6])

dot = np.dot(c,d)
dot
```

두 벡터의 내적이 0일 때, 두 벡터는 직교한다고 정의한다. 벡터 c, d의 내적은 42로, 0이 아니다. 따라서 두 벡터는 직교하지 않는다.

## 3. $|F|$ 를 구하세요. 이 데이터는 어떤 의미가 있나요?

\begin{align}
F = 
\begin{bmatrix}
    2 & -2 & 5 \\
    4 & 1 & 10 \\
    12 & 6 & 30
\end{bmatrix}
\end{align}

```
F = np.array([[2, -2, 5], [4, 1, 10], [12, 6, 30]])
F_det = np.linalg.det(F)
F_det
```

행렬 F의 행렬식은 값이 0 이다.

행렬식 값이 0이면 그 행렬의 역행렬은 존재하지 않는다는 것을 의미한다.

1. 행렬의 열벡터는 R_n 에서 독립이다.

2. 행렬의 행벡터는 R_n 에서 독립이다. (행렬의 벡터들이 선형 독립이다.)

3. 역행렬이 존재하지 않는다.

4. 기하학적으로 투영하면 차원이 줄어든다.

## 7. 아래 데이터에 대해 k-means clustering을 적용 후, 결과를 그래프로 그리세요.



```
points = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/sc13x/data.csv')
points.head()
```

```
points.shape
```

```
sns.scatterplot(x = 'x', y = 'y', data = points)
```

```
from sklearn.cluster import KMeans 
import warnings
warnings.filterwarnings(action='ignore')
```

```
sum_of_squared_distances = []
K = range(1, 15)
for k in K:
    km = KMeans(n_clusters = k)
    km = km.fit(points)
    sum_of_squared_distances.append(km.inertia_)
```

```
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10))
visualizer.fit(points)
```

```
km = KMeans(n_clusters = 4, n_jobs = 4, random_state=42)
km.fit(points)
points['cluster'] = km.labels_
points
```

```
sns.scatterplot(x="x", y="y", data=points, hue = 'cluster');
```

```
km2 = KMeans(n_clusters = 5, random_state=42)
km2.fit(points)
points['cluster2'] = km2.labels_
points
```

```
sns.scatterplot(x="x", y="y", data=points, hue = 'cluster2');
```

# Scree Plot

```
plt.title("Scree plot")
plt.xlabel("Number of Components")
plt.ylabel("Eigenvalue")
plt.plot(pca.explained_variance_, 'o-');
```

```
def screePlot(pca): 
    n = len(pca.explained_variance_)
    ind = np.arange(n)
    vals = pca.explained_variance_ratio_

    ax = plt.subplot()
    cumvals = np.cumsum(vals)   # 누적합 계산
    ax.bar(ind, vals)
    ax.plot(ind, cumvals, color = 'red')

    ax.set_xlabel('PC')
    ax.set_ylabel('variance')
    plt.title('Scree Plot')


screePlot(pca)
```

```
result = pca.explained_variance_ratio_                     
print(result)
print(result.cumsum())
```

# k값을 선택하는 과정

클러스터의 숫자를 설정할 때 다음의 방법론들을 이용할 수 있다.(위키피디아 참조)

1. Rule of thumb 으로 k의 값을 데이터의 수가 n 이라 할 때 n을 2로 나눈 값의 제곱근을 k로 사용할 수 있다.

2. Elbow Method 를 사용한다. 클러스터의 수를 순차적으로 늘려가면서 결과를 모니터링 한다. 만약 하나의 클러스터를 추가했을 때 이전보다 훨씬 더 나은 결과를 나타내지 않는다면, 이전의 클러스터의 수를 구하고자 하는 클러스터의 수로 설정한다.

3. 정보 기준 접근법 (Information Criterion Approach)
클러스터링 모델에 대해 가능도를 계산하는 것이 가능할 때 사용하는 방법으로, 베이지안 정보 기준 등이 있다. k-평균 클러스터링 모델의 경우 가우시안 혼합 모델에 가깝기 때문에, 가우시안 혼합 모델에 대한 가능도를 만들어 정보 기준 값을 설정할 수 있다.


이번 sprint challenge 문제에서는 세션에서도 배운 elbow plot을 사용하여 k 값을 선택하였다.

클러스터 간의 거리의 합이 급격히 떨어지는 구간의 k 값을 군집의 개수로 사용하는 알고리즘이며, 그래프의 기울기가 k = 4, 5에서 급격히 수평이 되는 것으로 보아 클러스터의 개수는 4 또는 5가 적절할 것이며 그 이상의 클러스터는 비용 대비 얻는 이득이 적을 것이라 판단하였다.

해당 k가 최적의 k인지는 모형 평가를 거쳐야 확인할 수 있으며,

 손실함수를 사용한 오차최소화 방법, 실루엣 점수, Dunn Index등 평가 방법에 따라 k의 값이 각각 다르게 나올 수 있다. 이 때는 상대를 설득하기 좋거나 논리적으로 더 타당한 값을 선택한다.

정리하자면 도메인 지식을 활용하고 그래프를 그려 '감'으로 k 값을 선택하는 것이 대표적이다.

하지만 '감'으로 선택한 행동을 설명하거나 남을 설득하기 어려울 때는 알고리즘의 도움을 받을 수 있다. 

yellow brick 이라는 패키지는 머신러닝 시각화를 지원해주는 패키지인데 이를 사용하면 그룹의 변동성을 확인함과 동시에 군집을 학습할 때 소요되는 시간까지 확인할 수 있어서 모델의 경제성도 어느정도 파악 가능하다.

결과적으로 elbow plot을 확인하면 k = 5 일 때 가장 inertia가 작으며, 그 이상 클러스터는 연산 속도가 급격히 증가하는 것으로 볼 때 최적의 k는 5라고 판단할 수 있다.
