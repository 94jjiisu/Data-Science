## Run a $\chi^{2}$ Test (Numpyë¡œ ì§ì ‘ ê³„ì‚°)

ê°€ëŠ¥í•œ 3ê°€ì§€ì˜ ì¡°í•© ì¤‘ 2ê°€ì§€ì˜ ê²½ìš°ì— ëŒ€í•´ì„œë§Œ ìˆ˜í–‰.

1) `income` & `pay_meal_out`

2) `pay_meal_out` & `persian_food`


````python
obs = pd.crosstab(small_food['income'], small_food['pay_meal_out'])
obs
````

- Numpyë¡œ ê³„ì‚°í•˜ê¸°ì—ëŠ” categoryê°€ ë„ˆë¬´ ë§ìœ¼ë‹ˆ 3, 4 ë°ì´í„°ë§Œ ì‚¬ìš©

| |pay = 3|pay = 4|
|:-:|:-:|:-:|
|income = 3|9|2|
|income = 4|13|3|


### ì˜ˆì¸¡ê°’ ê³„ì‚°
\begin{align}
expected_{i,j} =\frac{(row_{i} \text{total})(column_{j} \text{total}) }{(\text{total observations})}  
\end{align}

````python
obs = np.array([[9, 2], [13, 3]])
obs
````

````python
total_obs = np.sum(obs)
total_obs
````

````python
exp = np.array([[(9+2)*(9+13), (9+2)*(2+3)], [(13+3)*(9+13), (13+3)*(2+3)]])

exp = exp / total_obs
exp
````

### $\chi^{2}$ í†µê³„ì¹˜ ê³„ì‚°

\begin{align}
\chi^2 = \sum \frac{(observed_i-expected_i)^2}{(expected_i)}
\end{align}

````python
squared = np.power(obs - exp, 2)
squared
````

````python
x2 = np.sum(squared / exp)
x2
````

## 4. ìœ„ì—ì„œ ì‚¬ìš©í•œ one sample chisquare testë¥¼ í•¨ìˆ˜ì˜ í˜•íƒœë¡œ ë³€ê²½í•˜ë¼. 

````python
v1 = [18,22,20,15,23,22]
v2 = [5,23,26,19,24,23]

def myChisq(val):
  obs = val
  exp = np.sum(obs) / len(obs)
  chi_squared = ((obs-exp)**2) / exp
  chi = chi_squared.sum()
  p_value = 1 - stats.chi2.cdf(chi, df = len(val)-1)
  return print(f'ì£¼ì–´ì§„ ê°’ = {val}, ì¹´ì´ì œê³±ê°’ = {chi} , p-value = {p_value}') 


myChisq(v1) 
myChisq(v2)
````
